### Evaluate MLflow experiments

This is the most straightforward way to evaluate experiments.

However the metrics may differ slightly from the csv metrics. Hence, if high precision is required, you may want to evaluate the csv predictions instead.

```python
import sys
sys.path.append('..')
import napc.eval as ev 

# get all result df from mlflow
ev.setup_mlflow("../mlflow_creds/default.yaml")
run_df = ev.mlflow.search_runs(search_all_experiments=True).astype("string")
run_df.head()
```

```python
# shortcut to get runs for sepcific experiment by name
#run_df =ev.get_runs_from_experiment("2026-01-25-18-13-21_test")
```

```python
# calculate average performance for the experiment
metric_cols = run_df.filter(like="metrics.").columns.tolist()
run_df[metric_cols]=run_df[metric_cols].astype("float")
result_df = run_df.loc[:,metric_cols+["params.trainer.logger.experiment_name"]].groupby("params.trainer.logger.experiment_name").describe()
result_df
```

```python
# optional: preprocess for better column names and remove unneeded columns
df = ev.preprocess_df(run_df.copy(),metrics_mapping=ev.MLFLOW_METRICS,)
df.head()
```

```python
# optional: preprocess for better column names and remove unneeded columns
df = ev.preprocess_df(run_df.copy(),metrics_mapping=ev.MLFLOW_METRICS, long=True)
df.head()
```
