```python
import mlflow
import sys
sys.path.append('..')
from napc.helpers import setup_mlflow
setup_mlflow("../mlflow_creds/default.yaml")
```

```python
# Get all Running or Crashed runs in the experiment
NAME="2026-01-24-14-45-51_test" # enter experiment name here (al)

experiment = mlflow.get_experiment_by_name(NAME)
runs_df = mlflow.search_runs(experiment.experiment_id)
runs_df.head()
```

```python
runs = runs_df.query("status == 'RUNNING' or status == 'FAILED'")["run_id"].values.tolist()
# write run ids to a file
output_path = "../out/"+NAME+".txt"

with open(output_path, "w") as f:
    for run in runs:
        f.write(f"{run}\n")
```
